#### Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in real time?

Ans: Bokeh server makes it easy to create interactive web applications that connect front-end UI events to running Python code by converting the objects into JSOn format.

Now we can use the Bokeh server to create interactive plots that can be updated in real time by defining a Bokeh application that serves the plot and handles user interactions, and then running the Bokeh server to host the application.

Now we can see the plot update in real time, when changes made on bokeh interactive plot.
