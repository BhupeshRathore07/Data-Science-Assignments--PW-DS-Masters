#### Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

Ans: "LookupError" class exception in Python forms the base class for all exceptions that are raised when an index or a key is not found for a sequence or dictionary respectively.

```python
x = [1, 2, 3, 4]
try:
    print(x[10])
except LookupError as e:
    print(e)
```

```python
a = {'name': 'Lenin Mishra',
    'age': 28,
    'language': 'Python'
    }
inp = input('What do you want to search: ')

try:
    print(a[inp])
except LookupError as e:
    print("Key Error",e)
```
