#### Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

Ans: To build a URL to a specific function, use the url_for() function. It accepts the name of the function as its first argument and any number of keyword arguments.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Hello, World!'

@app.route('/about/<username>')
def about(username):
    return f'This is the about page for {username}'

with app.test_request_context():
    print(url_for('index'))
    print(url_for('about', username = "Bhupesh"))
```
