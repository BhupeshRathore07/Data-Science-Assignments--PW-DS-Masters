#### Q2. Write a Python function to estimate the population mean using a sample mean and standard deviation.

```python
import numpy as np

def estimate_pop_mean(sample_mean, sample_std, sample_size):
    pop_mean = sample_mean * (1.0 / (sample_size * sample_std ** 2))
    pop_mean /= (1.0 / (sample_size * sample_std ** 2))

    return pop_mean

sample_mean = 10
sample_std = 2
sample_size = 100

pop_mean = estimate_pop_mean(sample_mean, sample_std, sample_size)
print(f"Estimated population mean: {pop_mean}")
```

#### Q5. Write a Python script to conduct a hypothesis test on the difference between two population means, given a sample from each population.

```python
import numpy as np
from scipy.stats import t

sample1 = np.array([84, 81, 78, 75, 82, 79, 80, 83, 77, 81])
sample2 = np.array([72, 74, 68, 70, 73, 69, 71, 75, 68, 72])

mean1 = np.mean(sample1)
mean2 = np.mean(sample2)
std1 = np.std(sample1, ddof=1)
std2 = np.std(sample2, ddof=1)

# Calculate the test statistic and p-value
n1 = len(sample1)
n2 = len(sample2)
s_pooled = np.sqrt(((n1-1)*(std1**2) + (n2-1)*(std2**2))/(n1+n2-2))
t_stat = (mean1 - mean2)/(s_pooled*np.sqrt(1/n1 + 1/n2))
df = n1 + n2 - 2
p_value = 1 - t.cdf(t_stat, df)

alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

#### Q9. Generate a Student's t-distribution plot using Python's matplotlib library, with the degrees of freedom parameter set to 10.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import t

df = 10
x = np.linspace(t.ppf(0.001, df), t.ppf(0.999, df), 100)
y = t.pdf(x, df)

fig, ax = plt.subplots()
ax.plot(x, y, 'r-', lw=2, alpha=0.6, label='t pdf')
ax.set_title("Student's t-distribution (df=10)")
ax.set_xlabel('t-values')
ax.set_ylabel('Probability density')
ax.legend(loc='best', frameon=False)
plt.show()
```

#### Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two random samples of equal size and a null hypothesis that the population means are equal.

```python
import numpy as np
from scipy.stats import ttest_ind

sample1 = np.random.normal(5, 1, 100)
sample2 = np.random.normal(6, 1, 100)

t_statistic, p_value = ttest_ind(sample1, sample2)

print("t-statistic: ", t_statistic)
print("p-value: ", p_value)
```

#### Q12: What is t-statistic? State the formula for t-statistic.

Ans: The t-statistic is a measure of the difference between the sample mean and the population mean, relative to the variation within the sample. It is used in hypothesis testing to determine whether the difference between two sample means is statistically significant or not.


t = (x̄₁ - x̄₂) / [s²pooled / (n₁ + n₂ - 2)]^(1/2)

where:

x̄₁ and x̄₂ are the sample means of the two independent samples

s²pooled is the pooled estimate of the population variance (i.e., an estimate of the variance of the two populations combined)

n₁ and n₂ are the sample sizes of the two independent samples

#### Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.

Ans: Null hypothesis (H0): The true population mean difference is equal to 10 mmHg.

Alternative hypothesis (Ha): The true population mean difference is less than 10 mmHg.

The significance level is given as 0.05.

The formula for the t-statistic is:

t = (x̄ - μ) / (σ / sqrt(n))

t = (8 - 10) / (3 / sqrt(100))
= -2.82

For a one-tailed test at a significance level of 0.05 and 99 degrees of freedom, p-value is approximately 0.002.

Comparing the p-value with the significance level:

"We reject the null hypothesis."

The new drug does significantly decrease blood pressure by more than 10 mmHg.

#### Q16. Two groups of students are given different study materials to prepare for a test. Test the hypothesis that the population means for the two groups are equal with a significance level of 0.01.

Alternative hypothesis (Ha): The true population means for the two groups are not equal.

The formula for the t-statistic is:

t = (x̄1 - x̄2 - 0) / sqrt((σ1^2 / n1) + (σ2^2 / n2))

For a two-tailed test at a significance level of 0.01 and 68 degrees of freedom, the p-value is approximately 0.019.

Comparing the p-value with the significance level:
