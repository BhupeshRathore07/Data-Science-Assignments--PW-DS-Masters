#### Q1. Explain GET and POST methods.

Ans: GET Method:
The GET method is used to retrieve information from a server. When a client sends a GET request to a server, it includes a URL in the request. The server then responds with the requested information, which is typically displayed in the client's web browser

POST Method:
The POST method is used to submit data to a server for processing. When a client sends a POST request to a server, it includes data in the request body.

#### Q2. Why is request used in Flask?

Ans: The Request, in Flask, is an object that contains all the data sent from the Client to Server. This data can be recovered using the GET/POST Methods.

#### Q3. Why is redirect() used in Flask?

Ans: It returns a response object and redirects the user to another target location with specified status code where location parameter is the URL where response should be redirected.

#### Q4. What are templates in Flask? Why is the render_template() function used?

Ans: Templates are files that contain static data as well as placeholders for dynamic data.

render_template() is used to render the external HTML file to be returned as the response from the view function.
